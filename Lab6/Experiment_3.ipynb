```python
import numpy as np
import scipy
import scipy.optimize as opt
from scipy.interpolate import interp1d
import matplotlib
import math
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('trial1_cup.csv')
cup_time = pd.DataFrame.to_numpy(df, dtype=np.float32).transpose()[0]
cup_temp = pd.DataFrame.to_numpy(df, dtype=np.float32).transpose()[1]
plt.title('Temperature Change of Water in Styrofoam Cup')
plt.xlabel('Time (s)')
plt.ylabel('Temp (degrees C)')
plt.axvline(50, color='orange')
plt.axvline(100, color='orange')
plt.plot(cup_time, cup_temp)
```

```python
time1, time2 = cup_time[99], cup_time[199]
temp1, temp2 = cup_temp[99], cup_temp[199]
temp_avg = (temp1 + temp2) / 2 + 273.15
slope = (temp2 - temp1) / (time2 - time1)
print("Slope (degrees/s), dT_in/dt: ", np.round(slope,5))
print("Average temp (K), T_in: ", np.round(temp_avg))
```

```python
df = pd.read_csv('trial1_can.csv')
can_time = pd.DataFrame.to_numpy(df, dtype=np.float32).transpose()[0]
can_temp = pd.DataFrame.to_numpy(df, dtype=np.float32).transpose()[1]
plt.title('Temperature Change of Water in Metal Can')
plt.xlabel('Time (s)')
plt.ylabel('Temp (degrees C)')
plt.axvline(30, color='orange')
plt.axvline(60, color='orange')
plt.plot(can_time, can_temp)
```
